# 神经网络
神经网络的一个重要性质是它可以自动的从数据中国呢学习到合适的权重参数。
神经网络的组成包含输入层、隐藏层和输出层。
输入层被称为第0层，隐藏层和输出层被称为第1层、第2层，依此类推。
可以用权重层数来称呼神经网络，比如一个有两层权重的神经网络被称为两层神经网络。
也可以用层数来称呼神经网络，比如一个有两层权重的神经网络被称为三层神经网络。
引入新函数 h(x)，中文名为激活函数（activation function）。
激活函数的作用是对神经网络的输出进行非线性变换。
y = h(w1*x1 + w2*x2 + b)
常用的激活函数有阶跃函数、Sigmoid 函数、ReLU 函数等。
激活函数是连接感知机和神经网络的桥梁。

输入信号经过权重和偏置计算后得到 a，再经过激活函数得到输出 y。
感知机使用了阶跃函数作为激活函数。如果将激活函数换成其他函数，就得到了神经网络。

### sigmoid 函数
Sigmoid 函数的公式如下：
h(x) = 1 / (1 + exp(-x))
其中 exp 表示指数函数。

```python
import numpy as np
# 阶跃函数
def step_function(x): # x 可以是一个数，也可以是一个数组
    y = x > 0 # 此处得到的是一个 boolean 数组
    return y.astype(np.int_) # 转换为 int 数组，最终返回 0 和 1 组成的数组

# 对阶跃函数进行画图
import matplotlib.pylab as plt
a = np.arange(-5.0, 5.0, 0.1) # -5.0 到 5.0，步长为 0.1
b = step_function(a) # 调用 step_function 函数，得到 b
plt.plot(a, b) # 画图
plt.ylim(-0.1, 1.1) # y 轴的范围
plt.show() # 显示图像
```

```python
import numpy as np
# sigmoid 函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
# 对 sigmoid 函数进行画图
import matplotlib.pylab as plt
a = np.arange(-5.0, 5.0, 0.1) # -5.0 到 5.0，步长为 0.1
b = sigmoid(a) # 调用 step_function 函数，得到 b
plt.plot(a, b) # 画图
plt.ylim(-0.1, 1.1) # y 轴的范围
plt.show() # 显示图像
```

#### 阶跃函数和 sigmoid 函数都是非线性函数
#### 神经网络的激活函数必须是非线形函数！！！

## ReLU 函数（Rectified Linear Unit）
最近神经网络中使用的比较多的函数
它的公式如下：
h(x) = max(0, x)
即当 x 小于 0 时，h(x) 等于 0；当 x 大于等于 0 时，h(x) 等于 x。
**可以使用 numpy 的 maximum 函数**

```python
# Relu 函数
def relu(x):
    return np.maximum(0, x) # maximum 函数返回两个数中较大的一个
# 对 relu 函数进行画图
import matplotlib.pyplot as plt
a = np.arange(-5.0, 5.0, 0.1) # -5.0 到 5.0，步长为 0.1
b = relu(a) # 调用 relu 函数，得到 b
plt.plot(a, b)
plt.ylim(-0.1, 5)
plt.show()
```
